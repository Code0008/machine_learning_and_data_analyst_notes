```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

class Venta:
    def __init__(self, datos):
        """
        Inicializa la clase Venta con un DataFrame de datos.
        """
        self.datos = pd.DataFrame(datos)
    
    def agregar_fecha_vencimiento(self, fechas_vencimiento):
        """
        Agrega una columna de fecha de vencimiento en el formato dd/mm/AAAA.
        """
        self.datos['fecha_vencimiento'] = fechas_vencimiento
    
    def calcular_media_ponderada(self, columna_precio, columna_cantidad):
        """
        Calcula la media ponderada del precio de venta.
        """
        self.datos['media_ponderada'] = (self.datos[columna_precio] * self.datos[columna_cantidad]).sum() / self.datos[columna_cantidad].sum()
    
    def calcular_estadisticas(self, columna):
        """
        Calcula la mediana, moda, varianza y desviaciÃ³n estÃ¡ndar de una columna especÃ­fica.
        """
        mediana = self.datos[columna].median()
        moda = self.datos[columna].mode()[0]
        varianza = self.datos[columna].var()
        desviacion = self.datos[columna].std()
        return mediana, moda, varianza, desviacion
    
    def agregar_columna_vencido(self):
        """
        Agrega una columna indicando si el producto estÃ¡ vencido o no.
        """
        hoy = datetime.now()
        self.datos['vencido'] = self.datos['fecha_vencimiento'].apply(lambda x: 'SÃ­' if datetime.strptime(x, '%d/%m/%Y') < hoy else 'No')
    
    def calcular_rotacion_inventario(self):
        """
        Calcula la rotaciÃ³n de inventario para productos no vencidos.
        """
        self.datos['rotacion'] = (self.datos['precio_venta'] - self.datos['precio_compra']) / self.datos['stock_promedio']
    
    def ordenar_datos(self):
        """
        Ordena el DataFrame por ganancias de productos no vencidos y por categorÃ­a y stock.
        """
        self.datos = self.datos[self.datos['vencido'] == 'No']
        self.datos = self.datos.sort_values(by=['ganancia', 'stock', 'categoria'], ascending=[False, True, True])
    
    def mostrar_grafico_ventas_por_ubicacion(self):
        """
        Genera un grÃ¡fico de las ventas totales por ubicaciÃ³n y producto.
        """
        ventas_por_ubicacion = self.datos.groupby(['ubicacion', 'nombre_producto'])['precio_venta'].sum().unstack()
        ventas_por_ubicacion.plot(kind='bar', stacked=True, figsize=(10, 7))
        plt.title('Ventas Totales por UbicaciÃ³n y Producto')
        plt.xlabel('UbicaciÃ³n')
        plt.ylabel('Ventas Totales')
        plt.show()
    
    def mostrar_grafico_distribucion_ventas(self):
        """
        Genera un grÃ¡fico de la distribuciÃ³n de ventas por ubicaciÃ³n.
        """
        sns.histplot(data=self.datos, x='ubicacion', weights='precio_venta', kde=True)
        plt.title('DistribuciÃ³n de Ventas por UbicaciÃ³n')
        plt.xlabel('UbicaciÃ³n')
        plt.ylabel('Ventas')
        plt.show()
```
