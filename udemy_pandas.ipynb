```python
#convinar , merger, join
df1 = pd.DataFrame({
    "ID": np.arange(1,5),
    "Nombre":["Tilin", "Insano", "Insanin", "Elpepes"]
})

df2 = pd.DataFrame({
    "ID": np.arange(1,4),
    "Edad": np.random.randint(20, 30, (3,))
})

#convinamos ambos dataframes en uno creando uno nuevo en base a una columna en comun
#how = outer -> indicamos que los que no tengan data comun los agrege con nulos
#how = rigth
# how = left

df_convinado = pd.merge(df1, df2, on="ID", how="outer")
df_convinado
```

```python

df_indexado = pd.merge(df1, df2,  left_index=True, right_index=True)
df_indexado
```
